### Etapas

- Adição de Nós e Arestas: Mostramos como adicionar nós e arestas, tanto individualmente quanto em grupos.
- Atributos: Demonstramos como adicionar e manipular atributos tanto para nós quanto para arestas, o que é útil para armazenar informações adicionais como pesos, rótulos ou qualquer outra característica relevante.
- Métricas e Análise de Grafos: Calculamos e imprimimos métricas básicas que são fundamentais para a análise de redes, como graus de nós, caminhos mais curtos e centralidades.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# 1. Criação de Grafos
# Grafo simples não-direcionado
G = nx.Graph()
# Grafo direcionado
D = nx.DiGraph()
# Grafo multigrafo (permite múltiplas arestas entre os mesmos nós)
M = nx.MultiGraph()
```

```python
# 2. Adição de Nós e Arestas
# Adicionando nós
G.add_node("A")
G.add_nodes_from(["B", "C", "D", "E"])  # Adicionar múltiplos nós

# Adicionando arestas
G.add_edge("A", "B")
G.add_edges_from([("A", "C"), ("B", "D"), ("C", "E"), ("D", "E")])  # Adicionar múltiplas arestas
```

```python
# 3. Atributos
# Nós com atributos
G.nodes["A"]['role'] = "manager"
# Arestas com atributos
G.edges["A", "B"]['weight'] = 4.5
```

```python
# 4. Métricas e Análise de Grafos
# Graus dos nós
degrees = G.degree()
print("Graus dos Nós:", dict(degrees))
```

```python
# Caminho mais curto
path = nx.shortest_path(G, source="A", target="E")
print("Caminho mais curto de A para E:", path)
```

```python
# Centralidade de proximidade
closeness = nx.closeness_centrality(G)
print("Centralidade de proximidade:", closeness)
```

```python
# 5. Visualização
# Desenhando o grafo com Matplotlib
plt.figure(figsize=(8, 5))
pos = nx.spring_layout(G)  # Posicionamento dos nós
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, edge_color='gray', width=2)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, 'weight'))
plt.title("Visualização de Rede com NetworkX")
plt.show()
```
