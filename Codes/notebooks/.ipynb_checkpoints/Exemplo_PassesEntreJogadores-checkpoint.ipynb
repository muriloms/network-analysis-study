- Objetivo: Visualizar e analisar a rede de passes entre jogadores durante um jogo de futebol para entender as dinâmicas de jogo e identificar jogadores-chave.
- Dados: Criaremos dados fictícios para representar os passes entre jogadores de duas equipes.
- Análise: Usaremos métricas de rede para identificar os jogadores mais envolvidos nas jogadas e a visualização no campo para contextualizar a análise.

### Etapas:
- Definição das Equipes e Posições no Campo: Jogadores de duas equipes distintas são posicionados em metades diferentes do campo.
- Criação de um Grafo Direcionado com Pesos: O grafo agora inclui pesos que representam o número de vezes que um jogador passou a bola para outro.
- Visualização do Campo e dos Passes: O campo é dividido por uma linha central, e as arestas são desenhadas com espessuras variáveis para representar a quantidade de passes.
- Diferenciação das Equipes: Jogadores de cada equipe são coloridos de maneira diferente (azul e vermelho) para clara distinção

```python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
# Definindo os jogadores e suas posições no campo (coordenadas x, y)
team1 = {
    'Jogador 1': (10, 50), 'Jogador 2': (10, 20), 'Jogador 3': (10, 80),
    'Jogador 4': (30, 30), 'Jogador 5': (30, 70)
}
team2 = {
    'Jogador 6': (90, 50), 'Jogador 7': (90, 20), 'Jogador 8': (90, 80),
    'Jogador 9': (110, 30), 'Jogador 10': (110, 70)
}
```

```python
# Criando um grafo direcionado
G = nx.DiGraph()

# Adicionando nós (jogadores) ao grafo de forma correta
G.add_nodes_from(team1)
G.add_nodes_from(team2)
```

```python
# Adicionando arestas fictícias (passes) entre jogadores com pesos (número de passes)
passes_team1 = [('Jogador 1', 'Jogador 4', 5), ('Jogador 4', 'Jogador 5', 3),
                ('Jogador 5', 'Jogador 1', 2), ('Jogador 3', 'Jogador 4', 6)]
passes_team2 = [('Jogador 6', 'Jogador 7', 4), ('Jogador 7', 'Jogador 8', 5),
                ('Jogador 8', 'Jogador 9', 1), ('Jogador 9', 'Jogador 10', 3), ('Jogador 10', 'Jogador 6', 2)]


for src, dst, weight in passes_team1 + passes_team2:
    G.add_edge(src, dst, weight=weight)
```

```python
# Desenhando um campo de futebol
fig, ax = plt.subplots(figsize=(12, 7))
field = patches.Rectangle((0, 0), 120, 90, linewidth=2, edgecolor='green', facecolor='lightgreen')
midline = patches.Rectangle((60, 0), 1, 90, linewidth=0, edgecolor='white', facecolor='white')
ax.add_patch(field)
ax.add_patch(midline)

# Desenhando a rede de passes no campo
pos = {**team1, **team2}  # Combining team positions
edges = G.edges(data=True)
nx.draw_networkx_nodes(G, pos, node_color=['blue' if node in team1 else 'red' for node in G], node_size=700, alpha=0.8)
nx.draw_networkx_labels(G, pos, font_size=10, font_color='white', font_weight='bold', ax=ax)
nx.draw_networkx_edges(G, pos, arrowstyle='-|>', arrowsize=10, edge_color='gray',
                       width=[d['weight'] for *_, d in edges], ax=ax)

plt.xlim(0, 120)
plt.ylim(0, 90)
ax.set_xticks([])
ax.set_yticks([])
ax.set_title('Mapa de Passes entre Jogadores em um Jogo de Futebol (Dividido por Equipes)')
plt.axis('off')
plt.show()
```
